```
import numpy as np

def pa(arr):
  print('---')
  for row in arr:
    for val in row:
      print(val,' ',end='')
    print('')
  print('---')

#Input array
X=np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]])
#Output
y=np.array([[1],[1],[0]])

#Sigmoid Function
def sigmoid (x):
  return 1/(1 + np.exp(-x))

#Derivative of Sigmoid Function
def derivatives_sigmoid(x):
  return x * (1 - x)

#Variable initialization
epoch=1 #Setting training iterations
lr=0.1 #Setting learning rate
inputlayer_neurons = X.shape[1] #number of features in data set
hiddenlayer_neurons = 3 #number of hidden layers neurons
output_neurons = 1 #number of neurons at output layer

#weight and bias initialization
wh=np.around(np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons)),decimals=4)
bh=np.around(np.random.uniform(size=(1,hiddenlayer_neurons)),decimals=4)
wout=np.around(np.random.uniform(size=(hiddenlayer_neurons,output_neurons)),decimals=4)

bout=np.around(np.random.uniform(size=(1,output_neurons)),decimals=4)

pa(X)
pa(y)
print("==============e X,y=============")
pa(wh)
pa(bh)
pa(wout)
pa(bout)
print("==============e wh,bh,wout,bout=============")

for i in range(epoch):
  #Forward Propogation
  hidden_layer_input1=np.around(np.dot(X,wh),decimals=4)
  hidden_layer_input=np.around(hidden_layer_input1 + bh,decimals=4)
  hiddenlayer_activations = np.around(sigmoid(hidden_layer_input),decimals=4)
  output_layer_input1=np.around(np.dot(hiddenlayer_activations,wout),decimals=4)
  output_layer_input= np.around(output_layer_input1+ bout,decimals=4)
  output = np.around(sigmoid(output_layer_input),decimals=4)
  
  pa(hidden_layer_input)
  pa(hiddenlayer_activations)
  pa(output)
  print("==============e hidden input, hidden activations, output=============")

  #Backpropagation
  E = np.around(y-output,decimals=4)
  slope_output_layer = np.around(derivatives_sigmoid(output),decimals=4)
  slope_hidden_layer = np.around(derivatives_sigmoid(hiddenlayer_activations),decimals=4)
  d_output = np.around(E * slope_output_layer  ,decimals=4)
  Error_at_hidden_layer =np.around( d_output.dot(wout.T)  ,decimals=4)
  d_hiddenlayer = np.around(Error_at_hidden_layer * slope_hidden_layer,decimals=4)
  wout += np.around(hiddenlayer_activations.T.dot(d_output) *lr,decimals=4)
  bout += np.around(np.sum(d_output, axis=0,keepdims=True) *lr,decimals=4)
  wh += np.around(X.T.dot(d_hiddenlayer) *lr,decimals=4)
  bh += np.around(np.sum(d_hiddenlayer, axis=0,keepdims=True) *lr,decimals=4)
  
  pa(E)
  pa(slope_output_layer)
  pa(slope_hidden_layer)
  pa(d_output)
  pa(Error_at_hidden_layer)
  pa(d_hiddenlayer)
  print("==============e E slope at op, slope and hd, d_op, err at hd, d hd=============")
  pa(wout)
  pa(bout)
  pa(wh)
  pa(bh)
  print("==============e wout,bout,wh,bh=============")

```
